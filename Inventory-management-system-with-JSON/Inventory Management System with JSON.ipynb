```python
import json
import time

#initilizing variables
sales = " "

#loading the inventory records from JSON to string
fd = open("record.json","r")
js = fd.read()
fd.close()

#converting string records to JSON/dictionaries
record = json.loads(js)

#Displaying menu
print("--------------------------MENU-------------------------------\n")
for key in record.keys() :
    print(key,":", record[key]["Name"], "|" ,record[key]["Price"],"|", record[key]["Qn"])
print("-------------------------------------------------------------\n")    

#Taking user inputs regarding purchase and the user details
uin_name     = str(input("Enter your name             : "))
uin_mail     = str(input("Enter your mail             : "))
uin_phone    = int(input("Enter your phone number     : "))
uin_pro_id   = str(input("Enter the product id        : "))
uin_pro_quan = int(input("Enter the product quantity  : "))



#if we are having equal or more quantity than user wants
if (record[uin_pro_id]["Qn"] >= uin_pro_quan):
    
    #displaying purchase details
    print("-------------------------------------------------------------\n")
    print("-------------------------BILL--------------------------------\n") 
    print("Name of the product  : ", record [uin_pro_id]["Name"])
    print("price of the product : ", record [uin_pro_id]["Price"], " Rs")
    print("Quantity             : ", uin_pro_quan)
    print("-------------------------------------------------------------\n")
    print("Total Billing Amount : ",record[uin_pro_id]["Price"] * uin_pro_quan, " Rs")
    print("-------------------------------------------------------------\n")
    
    #updating the records dictionary
    record[uin_pro_id]["Qn"] =record[uin_pro_id]["Qn"] - uin_pro_quan
    
    #generating Sales structrue in csv
    sales = uin_name+","+uin_mail+","+str(uin_phone)+","+uin_pro_id+","+record[uin_pro_id]["Name"]
         +","+str(record [uin_pro_id]["Price"])+","+str(uin_pro_quan)+","+str(record[uin_pro_id]["Price"] * uin_pro_quan)
         +","+time.ctime() + "\n"

#if we are not having enough quantity
else:
    
    #displaying info regarding product quantity
    print("-------------------------------------------------------------\n")
    print("Sorry, we don't have the given quantity")
    print("We are only having ", record[uin_pro_id]["Qn"], "in stock")
    print("-------------------------------------------------------------\n")
    
    #taking user input for the purchase of whole quantity
    ch = input("Press y/Y to continue shopping: ")
    
    if (ch == "Y" or ch == "y"):
        
        #displaying purchase details
        print("-------------------------------------------------------------\n")
        print("-------------------------BILL--------------------------------\n") 
        print("Name of the product  : ", record [uin_pro_id]["Name"])
        print("Price of the product : ", record [uin_pro_id]["Price"] ," Rs")
        print("Quantity             : ",record [uin_pro_id]["Qn"])
        print("-------------------------------------------------------------\n")
        print("Total Billing Amount : ",record[uin_pro_id]["Price"] * record [uin_pro_id]["Qn"], " Rs")
        print("-------------------------------------------------------------\n")
        
        #generating sales structure in csv
        sales = uin_name+","+uin_mail+","+str(uin_phone)+","+uin_pro_id+","+record[uin_pro_id]["Name"]+","+str(record [uin_pro_id]["Price"])+","+str(record [uin_pro_id]["Qn"])+","+str(record[uin_pro_id]["Price"] * record [uin_pro_id]["Qn"])+","+time.ctime()+"\n"
        
        #updating the records dictionary
        record[uin_pro_id]["Qn"]= 0
        
    #if user has pressed something other than y    
    else:
        print("-------------------------------------------------------------\n")
        print("Thanks! Have a good day :)")

#converting dictionary to string
js= json.dumps(record)

#saving records in JSON file
fd = open("record.json", "w")
fd.write(js)
fd.close()

#saving records in sales file
fd = open("sales.txt","a")
fd.write(sales)
fd.close()

print("----------THANKS FOR YOUR ORDER, INVENTORY UPDATED!----------")
```
